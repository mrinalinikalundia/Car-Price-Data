```python
data=df[:10][['CarName','wheelbase','enginesize','horsepower']]
data
```

```python
Carname=data.iloc[:,0]
barwheelbase=data.iloc[:,1]
barengineratio=data.iloc[:,2]
barhorsepower=data.iloc[:,3]
index=np.arange(len(data))
```

```python
plt.figure(figsize=(20,10))
graphwheelbase=plt.bar(x=index,height=barwheelbase,width=0.35)
graphengineratio=plt.bar(x=index,height=barengineratio,width=0.35,bottom=barwheelbase)
graphhorsepower=plt.bar(x=index,height=barhorsepower,width=0.35,bottom=barwheelbase+barengineratio)
plt.legend(['wheelbase','engineratio','horsepower'])
plt.xlabel('CarName')
plt.ylabel('Features')
Carname=["\n".join(wrap(CarName,15))for CarName in Carname]
plt.xticks(index,Carname)
plt.title('Cars')
```
